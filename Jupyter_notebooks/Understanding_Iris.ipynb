```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from sklearn.datasets import load_iris
```

### Vamos a cargar el set de iris en numpy arrays y en pandas data frames para visualizarlo

```python
X, y = load_iris(return_X_y=True)
df, vf = load_iris(return_X_y=True, as_frame=True)
```

### Creamos un pandas data frame que contiene los atributos y las clases de cada flor

```python
af = df.copy()
af['class'] = vf.copy()
```

### Examinamos el data frame

```python
af.head(10)
```

### Vamos a hacer un plot de los datos utilizando matplotlib

```python
fig, ax = plt.subplots(4, 4, figsize = (13, 13))
dc = dict()
dc[0] = 'r'
dc[1] = 'y'
dc[2] = 'b'
for j, c1 in enumerate(df.columns.tolist()):
    for k, c2 in enumerate(df.columns.tolist()):
        ax[j][k].scatter(af[c1], af[c2], color = [dc[c] for c in af['class'].tolist()])
        ax[j][k].set_xlabel(c1)
        ax[j][k].set_ylabel(c2)
plt.show()
```

### Grabar los datos de pandas a un archivo
